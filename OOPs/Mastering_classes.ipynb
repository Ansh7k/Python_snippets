Right now, every robot we build will say the exact same thing. But objects usually have their own unique characteristics. For a robot, that might be a name, a model number, or a color. In Python, we store this unique data in variables attached to the object, called attributes.

The best way to give an object its attributes is the exact moment it's created. We do this using a special built-in method called the constructor, written as __init__ (with two underscores on each side). It runs automatically the second you instantiate the object.

```python
class House:
    def __init__(self, color):
        # 'self.color' is the attribute attached to the specific house
        # 'color' is the value we pass in when building it
        self.color = color

    def describe(self):
        # We use 'self.color' to access the data we saved earlier
        return f"This is a {self.color} house."

# We pass "blue" into the parentheses, which goes to the __init__ method
my_house = House("blue") 
print(my_house.describe())
```

In Python, inheritance allows a new class (the "child") to automatically gain all the attributes and methods of an existing class (the "parent"). This prevents us from rewriting the exact same code for concepts that share a foundation.

Think of it as an "is-a" relationship: A Drone is a Robot. It should be able to do everything a basic robot can do, plus its own unique actions.

To create a child class, we put the parent class name in parentheses when defining it

```python
class Robot: # class robot
    def __init__(self, name):
        self.name = name

    def say_hello(self):
        return f"Beep boop, I am {self.name}"

my_robot = Robot("R2-D2") # instance of class robot

print(my_robot.say_hello()) # instance calling the method of original class

class Drone(Robot): #inherit class 
    
    def fly(self):
        return "Whoosh! I am flying."
    
my_drone = Drone("Hoverbot")

print(my_drone.say_hello())
print(my_drone.fly())
```

Perfect execution. Your drone successfully inherited the name setup and the say_hello method from the base Robot blueprint, while also adding its own unique fly ability.

Right now, your Drone relies entirely on the Robot class's __init__ method to set itself up. But what if a drone needs specific data that a regular robot doesn't have, like a propeller_count?

If we just write a new __init__ inside Drone, it will completely overwrite (or "override") the parent's __init__, and the drone would forget how to set up its name.

To fix this, we use the super() function. It acts as a direct hotline to the parent class, allowing the child to say, "Hey parent, run your setup first, and then I'll do my specific setup."

Here is how that looks:

```python
class Drone(Robot):
    def __init__(self, name, propeller_count):
        # 1. Send the name up to the parent (Robot) to handle
        super().__init__(name) 
        
        # 2. Handle the Drone-specific data here
        self.propeller_count = propeller_count
```

```python
class Robot: # class robot
    def __init__(self, name):
        self.name = name

    def say_hello(self):
        return f"Beep boop, I am {self.name}"

my_robot = Robot("R2-D2") # instance of class robot

print(my_robot.say_hello()) # instance calling the method of original class

class Drone(Robot): #inherit class 
    def __init__(self, name, propeller_count):
        super().__init__(name)
        self.propeller_count = propeller_count

    def fly(self):
        return f"Whoosh! My {self.propeller_count} are spinning."
    
my_drone = Drone("Hoverbot", 4)

print(my_drone.say_hello())
print(my_drone.fly())
```

Let's look at object interaction 🤝.

The great thing about objects in Python is that they are treated just like any other piece of data (like a string or a number). This means you can pass an entire object as an argument directly into another object's method.

When one object interacts with another, the object performing the action is self, and the object receiving the action is passed in as a regular parameter.

Let's say we want R2-D2 to send a data packet to Hoverbot. Here is your mission:

Add a new method to your base Robot class called transmit_data.

It should take self as the first parameter, and a second parameter called target (which will represent the receiving object).

Inside the method, return an f-string that announces the transfer. You will use self.name to get the sender's name, and target.name to get the receiver's name (e.g., "R2-D2 sent a data packet to Hoverbot.").

At the very bottom of your script, test it by running: print(my_robot.transmit_data(my_drone))

```python
# everything we have build so far

class Robot: 
    # Class attribute shared by all robots
    population = 0 

    def __init__(self, name):
        self.name = name
        Robot.population += 1

    def transmit_data(self, target):
        # We access the name attribute directly from the target object
        return f"{self.name} sent a data packet to {target.name}."

    def say_hello(self):
        return f"Beep boop, I am {self.name}"

class Drone(Robot): 
    def __init__(self, name, propeller_count):
        # super() runs the parent's setup first
        super().__init__(name)
        self.propeller_count = propeller_count

    def fly(self):
        return f"Whoosh! My {self.propeller_count} propellers are spinning."
    
# --- Building our objects ---
my_robot = Robot("R2-D2") 
my_drone = Drone("Hoverbot", 4)

# --- Testing the interaction ---
print(my_robot.transmit_data(my_drone))

```
