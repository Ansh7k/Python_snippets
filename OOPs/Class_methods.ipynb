Let's explore class methods 🛠️.

So far, all the actions we've written (like say_hello or fly) have been instance methods. They take self as their first parameter because they are actions performed by an individual, specific robot.

But sometimes, an action belongs to the factory itself. For example, what if we want to ask the factory how many robots it has produced in total?

To do this, we use a class method. Instead of taking self (the object), it takes the class itself as the first parameter. By convention, we call this parameter cls (short for class).

We also have to flag it with a special marker called a decorator: @classmethod. Here is how it looks in our house blueprint:

```python
class House:
    zoning_type = "Residential" # Class attribute

    @classmethod
    def get_zoning(cls):
        # We use 'cls' instead of 'self' to access class-level data
        return f"All houses built by this blueprint are {cls.zoning_type}."

# We call the method on the Class itself, not an instance!
print(House.get_zoning())
```
